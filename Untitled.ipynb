```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
df = pd.read_csv('./archive/smoking_driking_dataset_Ver01.csv')
df.head()
```

```python
def age_cat(age):
    if age >= 20 and age <= 35:
        return 'AGE: 20-35'
    elif age >= 36 and age <= 45:
        return 'AGE: 36-45'
    elif age >= 46 and age <= 60:
        return 'AGE: 46-60'
    else:
        return 'AGE: 61-85'

def height_cat(h):
    if h >= 130 and h <= 155:
        return 'Height: 130-155'
    elif h >= 156 and h <= 160:
        return 'Height: 156-160'
    elif h >= 161 and h <= 170:
        return 'Height: 161-170'
    else:
        return 'Height: 171-190'
    
def weight_cat(d):
    if d >= 25 and d <= 55:
        return 'Weight: 25-55'
    elif d >= 56 and d <= 60:
        return 'Weight: 56-60'
    elif d >= 61 and d <= 70:
        return 'Weight: 61-70'
    else:
        return 'Weight: 71-140'   

def waistline_cat(d):
    if d >= 8 and d <= 75:
        return 'Waistline: 8-75'
    elif d >= 76 and d <= 81:
        return 'Waistline: 76-81'
    elif d >= 82 and d <= 88:
        return 'Waistline: 82-88'
    else:
        return 'Waistline: >88'   
    
def sightr_cat(d):
    if d >= 0.1 and d <= 0.7:
        return 'Sight right: 0.1-0.7'
    elif d >= 0.71 and d <= 1.0:
        return 'Sight right: 0.71-1.0'
    elif d >= 1.1 and d <= 1.2:
        return 'Sight right: 1.1-1.2'
    else:
        return 'Sight right: >1.2'   

def sightl_cat(d):
    if d >= 0.1 and d <= 0.7:
        return 'Sight left: 0.1-0.7'
    elif d >= 0.71 and d <= 1.0:
        return 'Sight left: 0.71-1.0'
    elif d >= 1.1 and d <= 1.2:
        return 'Sight left: 1.1-1.2'
    else:
        return 'Sight left: >1.2'   
    
    
def sbp_cat(d):
    if d >= 67 and d <= 112:
        return 'SBP: 67-112'
    elif d >= 68 and d <= 120:
        return 'SBP: 68-120'
    elif d >= 121 and d <= 131:
        return 'SBP: 121-131'
    else:
        return 'SBP: >132'   

def dbp_cat(d):
    if d >= 32 and d <= 70:
        return 'DBP: 32-70'
    elif d >= 71 and d <= 76:
        return 'DBP: 71-766'
    elif d >= 77 and d <= 82:
        return 'DBP: 77-82'
    else:
        return 'DBP: >83'
    
def hdl_cat(d):
    if d >= 1 and d <= 46:
        return '1-46'
    elif d >= 47 and d <= 55:
        return '47-55'
    elif d >= 56 and d <= 66:
        return '56-66'
    else:
        return '>67' 
    
def ldl_cat(d):
    if d >= 1 and d <= 89:
        return 'HDL: 1-89'
    elif d >= 90 and d <= 111:
        return 'HDL: 90-111'
    elif d >= 112 and d <= 135:
        return 'HDL: 112-135'
    else:
        return 'HDL: >136' 
    
def tri_cat(d):
    if d >= 1 and d <= 73:
        return 'Tri: 1-73'
    elif d >= 74 and d <= 106:
        return 'Tri: 74-106'
    elif d >= 107 and d <= 159:
        return 'Tri: 107-159'
    else:
        return 'Tri: >160' 
    
def hemo_cat(d):
    if d >= 1 and d <= 13.2:
        return 'hemo: 1-13.2'
    elif d >= 13.3 and d <= 14.3:
        return 'hemo: 13.3-14.3'
    elif d >= 14.4 and d <= 15.4:
        return 'hemo: 14.4-15.4'
    else:
        return 'hemo: >15.5' 
    
def urine_cat(d):
    if d == 6:
        return 'Urine: 6'
    else:
        return 'Urine: < 6'
    
def serum_cat(d):
    if d >= 0.1 and d <= 0.7:
        return 'Serum: 0.1-0.7'
    elif d >= 0.71 and d <= 0.8:
        return 'Serum: 0.71-0.8'
    elif d >= 0.81 and d <= 1.0:
        return 'Serum: 0.81-1.0'
    else:
        return 'Serum: >1.1' 
    
def ast_cat(d):
    if d >= 1 and d <= 19:
        return 'Ast: 1-19'
    elif d >= 20 and d <= 23:
        return 'Ast: 20-23'
    elif d >= 24 and d <= 28:
        return 'Ast: 24-28'
    else:
        return 'Ast: >29' 
    
def alt_cat(d):
    if d >= 1 and d <= 15:
        return 'Alt: 1-15'
    elif d >= 16 and d <= 20:
        return 'Alt: 16-20'
    elif d >= 21 and d <= 29:
        return 'Alt: 21-29'
    else:
        return 'Alt: >30'
    
def gamma_cat(d):
    if d >= 1 and d <= 16:
        return 'Gamma: 1-16'
    elif d >= 17 and d <= 23:
        return 'Gamma: 17-123'
    elif d >= 24 and d <= 39:
        return 'Gamma: 24-39'
    else:
        return 'Gamma: >40' 
    
def blds_cat(d):
    if d >= 25 and d <= 88:
        return 'Blds: 25-88'
    elif d >= 89 and d <= 96:
        return 'Blds: 89-96'
    elif d >= 97 and d <= 105:
        return 'Blds: 97-105'
    else:
        return 'Blds: >106' 
    
def tot_cat(d):
    if d >= 30 and d <= 169:
        return 'Tot: 30-169'
    elif d >= 170 and d <= 193:
        return 'Tot: 170-193'
    elif d >= 194 and d <= 219:
        return 'Tot: 194-219'
    else:
        return 'Tot: >220' 
```

```python
print('df shape:', df.shape)
print(df.describe())
```

```python
data = []
for i in tqdm(range(len(df))):
    d = [df.at[i,'sex'], age_cat(df.at[i, 'age']), height_cat(df.at[i, 'height']),
         weight_cat(df.at[i, 'weight']), waistline_cat(df.at[i, 'waistline']), sightl_cat(df.at[i, 'sight_left']), 
         sightr_cat(df.at[i, 'sight_right']), 'hear left'+str(df.at[i, 'hear_left']), 'hear right'+str(df.at[i, 'hear_right']),
         sbp_cat(df.at[i, 'SBP']), dbp_cat(df.at[i, 'DBP']), blds_cat(df.at[i, 'BLDS']), tot_cat(df.at[i, 'tot_chole']), hdl_cat(df.at[i, 'HDL_chole']),
         ldl_cat(df.at[i, 'LDL_chole']), tri_cat(df.at[i, 'triglyceride']), hemo_cat(df.at[i, 'hemoglobin']),
         urine_cat(df.at[i, 'urine_protein']), serum_cat(df.at[i, 'serum_creatinine']),
         ast_cat(df.at[i, 'SGOT_AST']), alt_cat(df.at[i, 'SGOT_ALT']), gamma_cat(df.at[i, 'gamma_GTP']),
         'smk'+str(df.at[i, 'SMK_stat_type_cd']), 'drk'+df.at[i, 'DRK_YN']]
    data.append(d)
```
